## Essential Shortcuts
Here's the ones I use the most. 

1. Run a cell "shift+enter"
2. Go up or down with selected cell 
- "Esc" then press "k" to go UP as many times as you want  
- Enter will end command mode and take you into a cell   
- "Esc" then press "j" as many times as you want     
3. Cut and paste a cell
- Cut. "Esc" to enter command mode then x or can use shift j or k to select multiple consecutive cells. 
- Paste press v
- I often use cut to just delete cells but you can also use vim's command Esc followed by "dd"
4. Copy is just like cut but use "c" instead of v
5. New cell below: Esc followed by "b"
6. New cell above: Esc followed by "a"
7. Save "Ctrl + S" Do this often!! Jupyter can crash when you least suspect it   
8. Change cell type to markdown: "Escape+m"
9. Change cell type to code: "Escape+y"

# Practical example

# Cool! Now what?
So we've created the functionality to extract the price from coin market cap. The next step is to make it into a function we can use in other places in code. It would also be cool to see if it works for bitcoin prices too

```python
def get_coin_market_cap_price(url):
    r=requests.get(url)
    html=r.text
    soup=bs4.BeautifulSoup(html)
    price_divs=soup.findAll("div",{"class": "priceValue"})
    price_div=price_divs[0]
    price=float(price_div.text.replace("$",""))
    return price
base_url="https://coinmarketcap.com/currencies/"
for currency in ["bitcoin","dogecoin"]:
    print(get_coin_market_cap_price(f"{base_url}{currency}/"))
```

```python
#kinda works, notice we got an error with "," at first but fixed it below. Bitcoins are worth a lot more than doge
#Nice we used Jupyter because we could quickly identify the error and test fixes without losing our coding "session"
```

```python
def get_coin_market_cap_price(url):
    r=requests.get(url)
    html=r.text
    soup=bs4.BeautifulSoup(html)
    price_divs=soup.findAll("div",{"class": "priceValue"})
    price_div=price_divs[0]
    price=float(price_div.text.replace("$","").replace(",",""))
    return price
base_url="https://coinmarketcap.com/currencies/"
for currency in ["bitcoin","dogecoin"]:
    print(get_coin_market_cap_price(f"{base_url}{currency}/"))
```

```python
#We can store the data a little better too
rows=[]
for currency in ["bitcoin","dogecoin"]:
    row={"currency": currency, "price": get_coin_market_cap_price(f"{base_url}{currency}/")}
    rows.append(row)
rows
```

```python
import pandas as pd#data manipulation library
current_price_df=pd.DataFrame(rows)
current_price_df
```

```python
current_price_df['price'].max()
```

```python
current_price_df['price'].describe()
```

```python
import datetime
current_price_df['datetime']=datetime.datetime.utcnow()
current_price_df
```

```python
current_price_df.to_csv("prices.csv",index=False)
```
